### Problem 1
In this question, we will use the csv file **Temp.csv**. 
- Read it into a data frame, call it `Temp`.

```python
import pandas as pd

Temp = pd.read_csv('./Temp.csv')
Temp.head()
```

- We learned that the temperature at New Haven, Connecticut is currently 58 degree fahrenheit. Add this information into the DataFrame.

```python
Temp.loc[-1] = ['New Haven', 'Connecticut', 58]
Temp.index += 1
Temp = Temp.sort_index()
```

```python
Temp
```

- Replace the names of the states by their abbreviations. That is 
    - Connecticut --> CT
    - New Jersey  --> NJ
    - New York    --> NY

```python
state_dict = {'Connecticut': 'CT', 'New Jersey': 'NJ', 'New York': 'NY'}
state_list = []
for state in Temp.State:
    state_list.append(state_dict[state])
Temp.State = state_list
Temp
```

- Select the rows in New York.

```python
NY_rows = Temp[Temp.State == 'NY']
NY_rows
```

- What is the average of the temperature of cities in NY?

```python
NY_rows.Temperature.mean()
```

- Find the mean and the standard deviation of the temperature of each state.

```python
Temp.groupby('State').agg(['mean', 'std'])
```

- We know the capital of each state: 
    - Connecticut --> Hartford
    - New Jersey  --> Treton
    - New York    --> Albany
    
    Create a data frame, `Capital`, with two columns: State and Capital, which should look like:
    
|| State| Capital|
| :-: |:---: | :---:|
| 0| CT  | Hartford|
| 1| NJ | Treton|
| 2| NY| Albany|

```python
Capital = pd.DataFrame({'State': ['CT', 'NJ', 'NY'], 'Capital': ['Hartford', 'Treton', 'Albany']})
Capital
```

- Mutate a new boolean column, `Capital`, to `Temp`, which is `True` if the city in the row is the capital of the state, else `False`.

```python
Temp = Temp.merge(Capital, how = 'outer', on = 'State')

Temp.Capital = (Temp.City == Temp.Capital)
Temp
```

- a. Load the Gross Domestic Product data for the 190 ranked countries in the file **GDP.csv**. Load **Country.csv** into a DataFrame.

```python
#### Your code here
# (a)
import numpy as np
import pandas as pd
gdp = pd.read_csv('GDP.csv', skiprows=range(5), header= None, usecols=[0,1,3,4], nrows=190, encoding='ISO-8859-1')
gdp.columns = ['CountryCode', 'Rank', 'CountryName', 'GDP']
country = pd.read_csv('Country.csv', encoding='ISO-8859-1')
```

```python
# (b)
combined = country.merge(gdp, how = 'inner', on = 'CountryCode')
len(country.merge(gdp, how = 'inner', on = 'CountryCode'))
combined.head()
```

```python
# (c)
combined = combined.sort_values(by='Rank', ascending = False)
combined.loc[12]
```

```python
# (d)
temp_list = []
for amt in combined['GDP']:
    temp_list.append(int(''.join(amt.split(','))))

combined['GDP'] = temp_list
```

```python
combined[['Income Group', 'GDP']].groupby('Income Group').mean()[:2]
```
